```python
# Entropy (discrete)
def H(probs):
    entropy = 0
    for p in probs:
        entropy -= np.log2(p) * p #remember that log(1/p) = -log(p)
    return entropy

# Example probabilities (fiddle with it to experiment)
probabilities = [.4, .2, .2, .2]

# Calculate
print('Entropy of p(x): ', H(probabilities))
```

```python
# Solutions from practice
pxy = np.array([[.1, .2],
                [.4, .1],
                [.1, .1]])

px = np.sum(pxy, axis = 0) #column-wise sum
py = np.sum(pxy, axis = 1) #row-wise sum

# Entropy of x
H(px)
```

```python
# Entropy of y
H(py)
```

```python
# Cross Entropy (discrete)
def xH(p,q):
    if len(p) != len(q):
        return 'p and q different lengths'

    crossentropy = 0
    for i in range(len(p)):
        crossentropy -= np.log2(q[i]) * p[i]

    return crossentropy

# For Y|X=0 and Y|X=1
pyx0 = pxy[:,0]/np.sum(pxy[:,0])
pyx1 = pxy[:,1]/np.sum(pxy[:,1])

xH(pyx0, pyx1)
```

```python
# KL Divergence (discrete)
def KLD(p, q):
    if len(p) != len(q):
        return 'p and q different lengths'

    divergence = 0
    for i in range(len(p)):
        divergence += p[i]*np.log2(p[i]/q[i])

    return divergence

KLD(pyx0, pyx1)
```

```python
# Verify it's xH - H
xH(pyx0,pyx1) - H(pyx0)
```

```python
# Mutual Information (discrete)
def MI(pxy):
    # get the marginal distributions
    px = np.sum(pxy, axis = 0)
    py = np.sum(pxy, axis = 1)

    # the outer product gets the p(x)p(y) values
    ind_probs = np.outer(py, px) # careful with order here

    # take advantage of broadcasting arrays
    return np.sum(pxy * np.log2(pxy / ind_probs))

MI(pxy)
```
