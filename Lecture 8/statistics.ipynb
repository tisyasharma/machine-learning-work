```python
from scipy.integrate import quad
import numpy as np

def p(x):
    return 2 * np.exp(-2 * x)

def xp(x):
    return x*p(x)

# check that the pdf is valid (the sum of probabilities is 1)
I, err = quad(p, 0, np.inf)
print(I)
```

```python
pxl5, err = quad(p, 0, 2)
pxl5
```

```python
# get the expected value
Ex, err = quad(xp, 0, np.inf)
print(Ex)
```

```python
# calculating population variance in python
X = np.array([1, 2, 3])

print(f'NumPy .var function: {np.var(X)}')
print(f'NumPy .var function with ddof arg: {np.var(X, ddof=1)}')
# The ddof=1 argument is used when calculating the sample variance in Python,
# specifically with numpy.var() or numpy.std(). ddof means divide by N-1
print(f'Population variance: {np.sum((X - X.mean())**2)/len(X)}')
print(f'Sample variance: {np.sum((X - X.mean())**2)/(len(X)-1)}')
print(f'NumPy .std function: {np.std(X)}')
print(f'NumPy .std function with ddof arg: {np.std(X, ddof=1)}')
print(f'Sample standard deviation: {np.sqrt(np.sum((X - X.mean())**2)/(len(X)-1))}')
```
