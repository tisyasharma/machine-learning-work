The approach below is called Leave-One-Out Cross-Validation (LOO-CV) with KNN:

Instead of predicting a single new point, it removes each row one by one.

Uses the remaining data to predict its label.

This effectively validates KNN across the entire dataset.


```python
# this function assumes that the first column is the label
# and that the rest of the columns are the features used to predict
# also assumes using the L2 norm
def knn(data, k = 'Default'):

    if k == 'Default':
        k = round(np.sqrt(data.shape[0]))
    else:
        k = k

    pred_labels = []
    for i in range(data.shape[0]):
        data_t = data.T
        studenti = data_t.pop(i).to_numpy()[1:] # remove element i from dataset
        rest = data_t.T.to_numpy() # dataset now lacks element i

        dist_i = []
        for j in range(rest.shape[0]):
            studentj = rest[j,1:]
            dist_i.append(np.linalg.norm(studenti - studentj))

        knns = np.where(np.argsort(dist_i)<k)[0]
        pred_labels.append(st.mode(rest[knns,0]))

    true_labels = data.to_numpy()[:,0].T

    return (np.array(pred_labels), true_labels)

coop_knn = knn(knn_df)
```

```python
preds, truth = coop_knn
```

```python
# One way of assessing performance
final_results = list(zip(truth, preds))
Counter(final_results)
```
