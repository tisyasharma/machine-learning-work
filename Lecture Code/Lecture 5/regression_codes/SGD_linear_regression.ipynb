```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assume the data below is your input data.
# Solve the problem using linear gression
# and the closed-formed solution

# Features (5 samples, 1 feature)
x = np.array([[1],
              [2],
              [3],
              [4],
              [5]])

# Targets
y = np.array([[3],
              [5],
              [7],
              [9],
              [11]])
```

```python
# The first thing is to define the phi matrix and the initial weights
phi = np.array([[1,1],
              [2,1],
              [3,1],
              [4,1],
              [5,1]])

y = np.array([[3],
              [5],
              [7],
              [9],
              [11]])

w = np.array([[1], [1]])
```

```python
# Let's give the user the chance to enter the learning rate and the
# number of iterations
lr = float(input('Enter the learning rate\n'))
niter = int(input('Enter the number of iterations\n'))
```

```python
# --- sklearn: Gradient Descent with SGDRegressor (linear case) ---
from sklearn.linear_model import SGDRegressor
import numpy as np

# Use existing phi and y. phi has columns [x, 1]; let sklearn learn the intercept.
X = phi[:, :1]            # only the feature column
y_vec = y.flatten()         # sklearn expects 1D target

# Reuse your inputs for learning rate and iterations
# (collected earlier as lr and niter)
sgd = SGDRegressor(
    loss="squared_error",
    penalty=None,          # no regularization; pure GD on MSE
    alpha=0.0,
    learning_rate="constant",
    eta0=lr,               # step size
    max_iter=niter,
    tol=None,              # run full niter
    random_state=0
)

sgd.fit(X, y_vec)

# Coefficients and intercept
coef_x = sgd.coef_[0]
intercept = sgd.intercept_[0]
print(f"SGD learned: y ≈ {coef_x:.6f} * x + {intercept:.6f}")

# Predictions for your downstream plot (overwrite to compare SGD)
final_predictions = sgd.predict(X).reshape(-1, 1)
print('First 5 predictions (SGD):', final_predictions[:5].ravel())
```
