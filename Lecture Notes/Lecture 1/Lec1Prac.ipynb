```python
import autograd.numpy as np
from autograd.numpy import log
from autograd.numpy import exp
from autograd import grad

from scipy.integrate import quad
```

```python
def f(x):
    return 1*x + 2*x + 3*x

# Create derivative function
df = grad(f)

# Evaluate derivative at a point, say x = 5
print(df(5.0))  # should print 6.0
```

```python
def g(x):
    return x**2 - x + 1

Area, error= quad(g, 0, 1)
print("Integral Value: ", Area)
print("Estimated Error: ",  error)
```

```python
def u(x):
    return (2*x + 1)*np.exp(-2*x) - np.log(x**2)/np.log(2)

                    
# Create derivative function
df = grad(u)

# Evaluate derivative at a point, say x = 5
print(df(5.0)) 
```
