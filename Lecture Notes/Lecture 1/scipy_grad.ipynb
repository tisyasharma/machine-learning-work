```python
# import the packages to perform integrals
from scipy.integrate import quad 
```

```python
# Function to define the original function
def f(x):
    return 2*x

# quad(f,0,2) computes the integral of f between 0 and 2.
# When we write Area, err = quad(f,0,2), it means that the value of the integral
# will be stored in the variable Area, whereas the value of the error will be stored
# in the variable err. Notice that later we did not use err, so we might as well written
# Area = quad(f,0,2). Try this and see what you get (perhaps a tuple?)
Area, err = quad(f, 0, 2)
print('Area = %.2f'%Area)
```
